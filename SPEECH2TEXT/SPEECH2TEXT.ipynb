```python
import torch
import sounddevice as sd
from transformers import Wav2Vec2ForCTC, Wav2Vec2Processor
```

```python
# Load pre-trained model and processor
processor = Wav2Vec2Processor.from_pretrained("facebook/wav2vec2-base-960h")
model = Wav2Vec2ForCTC.from_pretrained("facebook/wav2vec2-base-960h")
```

```python
# Define callback function for the stream
def callback(indata, frames, time, status):
    try:
        # Convert numpy array to tensor
        waveform = torch.from_numpy(indata)

        # Preprocess audio
        input_values = processor(waveform, sampling_rate=16000, return_tensors="pt").input_values

        # Perform inference
        with torch.no_grad():
            logits = model(input_values).logits

        # Decode logits to text
        predicted_ids = torch.argmax(logits, dim=-1)
        transcription = processor.decode(predicted_ids[0])

        print("Transcription: ", transcription)
    except Exception as e:
        print(f"Error: {e}")
```

```python
# Create stream
stream = sd.InputStream(callback=callback, channels=1, samplerate=16000)
```

```python
# Start stream
print("Starting transcription. Press Ctrl+C to stop.")
try:
    with stream:
        while True:  # Keep the stream open until interrupted
            sd.sleep(1000)  # Sleep for 1 second at a time
except KeyboardInterrupt:
    print("Transcription stopped.")
```
